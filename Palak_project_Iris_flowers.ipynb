```python
import sys
import scipy
import numpy
import matplotlib
import pandas
import sklearn
from pandas import read_csv
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.naive_bayes import GaussianNB
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

names=['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = read_csv('datasets_2.txt', names=names)
dataset.to_csv('datasets_2.csv',index=None)
print(dataset.shape)
print(dataset.head(20))
print(dataset.groupby('class').size())
X=dataset.values[:,0:4]
Y=dataset.values[:,4]
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=1)
G=GaussianNB()
G.fit(X_train,Y_train)
y_predictions=G.predict(X_test)
print('Accuracy Score:',accuracy_score(Y_test,y_predictions)*100)
print('confusion_matrix:',confusion_matrix(Y_test,y_predictions))
print('classification_report:',classification_report(Y_test,y_predictions))
clf = DecisionTreeClassifier(max_depth = 2, random_state = 0)
clf.fit(X_train,Y_train)
Y_pred=clf.predict(X_test)
print('Accuracy Score:',accuracy_score(Y_test,Y_pred)*100)
print('confusion_matrix:',confusion_matrix(Y_test,Y_pred))
print('classification_report:',classification_report(Y_test,Y_pred))
#tree.plot_tree(clf)
tree.plot_tree(clf,feature_names = names[0:4], class_names=['Iris-setosa','Iris-versicolor','Iris-virginica'],filled = True);
```
